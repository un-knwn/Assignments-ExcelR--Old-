```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python
#Reading & Storing the data 
Salary_Train = pd.read_csv("SalaryData_Train.csv")
Salary_Test = pd.read_csv("SalaryData_Test.csv")
```

```python
Salary_Train["Salary"].value_counts()
```

```python
Salary_Train.head()
```

```python
Salary_Test.head()
```

```python
# Creating a list for categorical data
string_columns=['workclass','education','maritalstatus','occupation','relationship','race','sex','native','Salary']
```

```python
# Labelling

number = LabelEncoder()
for i in string_columns:
    Salary_Train[i]= number.fit_transform(Salary_Train[i])
    Salary_Test[i]=number.fit_transform(Salary_Test[i])
```

```python
Salary_Train.head()
```

```python
Salary_Test
```

```python
#Copying Salary_Test values by droppping Salary column i.e target
Salary_Test1=Salary_Test.drop('Salary',axis=1)
Salary_Test1
```

```python
# Assigning X & Y values
X_train=Salary_Train.iloc[:,0:13]
y_train=Salary_Train.iloc[:,13]
#X_train = Salary_Train.drop('Salary', axis=1)
#y_train = Salary_Train['Salary']
```

# Multinomial Naive Bayes

```python
# Create a Multinomial Naive Bayes classifier
classifier = MultinomialNB()
```

```python
# Fit the classifier to the training data
classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = classifier.predict(Salary_Test1)
y_pred
```

```python
# Evaluate the classifier's performance
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_pred, Salary_Test['Salary'])
print("Accuracy:", accuracy)
```
