```python
# Applying T-test as there are two units to test & Data is normally distributed
# A two-tailed t-test is used to check whether the relations between variables are in any direction or not.
# Given Significane 5% i.e alpha = 0.05
# For two-tailed test alpha = 0.05/2 = 0.025

# Hypothesis testing statements
# H0 = There is no difference in diameter of cutlet
# H1 = There is difference in diameter of cutlet
```

```python
import pandas as pd
import scipy as sc
import matplotlib.pyplot as plt
import statsmodels.api as sm          
```

```python
#importing file
Cutlets = pd.read_csv("cutlets.csv")
Cutlets.head()
```

```python
#Checking for null values in data
Cutlets.isnull().sum()
```

```python
#checking normal distribution or not by  using qqplot
plt.figure(figsize = (5,4))
sm.qqplot(Cutlets["Unit A"], line = 'q')
plt.title('Unit A')
sm.qqplot(Cutlets["Unit B"], line = 'q')
plt.title('Unit B')
plt.show()
```

```python
#Its also for checking normal dist     pvalue shld be > 0.05
print(sc.stats.shapiro(Cutlets['Unit A']))
print(sc.stats.shapiro(Cutlets['Unit B']))
```

```python
#T-test 
statistic , p_value = sc.stats.ttest_ind(Cutlets['Unit A'],Cutlets['Unit B'], alternative = 'two-sided')    #ttest_ind = 2 independent sample
print('p_value=',p_value)
```

```python
alpha = 0.5/2
H0 = "There is no difference in diameter of cutlet"
H1 = 'There is difference in diameter of cutlet'
```

```python
if p_value <= alpha:
    print( 'we reject null hypothesis & ',H1)
else:
    print('we accept null hypothesis &',H0)
```
