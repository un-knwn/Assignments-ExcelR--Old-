# Pivot

```python

BookPivot = Book.pivot(index='Unnamed: 0',
                                 columns='BookTitle',
                                 values='BookRating')
BookPivot
```

```python
#Impute those NaNs with 0 values
BookPivot.fillna(0, inplace=True)
BookPivot
```

# Calculating Cosine Similarity between Users

```python

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1-pairwise_distances(BookPivot.values,metric='cosine')      #1-to get similarities
user_sim
```

```python
user_sim.shape
```

```python
#Store the results in a dataframe
user_sim1 = pd.DataFrame(user_sim)
user_sim1.iloc[0:20,0:20]
```

```python
#Set the index and column names to user ids 
user_sim1.index = Book.UserID
user_sim1.columns = Book.UserID
user_sim1
```

```python
#Filling Diagonal  with 0
np.fill_diagonal(user_sim, 0)
user_sim1.iloc[0:15, 0:15]
```

```python
#Most Similar Users
user_sim1.idxmax(axis=1)[0:50]   #idxmax = gives max in row, axis =1 scan clmnwise
```

```python
def recommend(custID):
    simID = user_sim1.loc[custID].idxmax()
    simID_books = Book[Book['UserID'] == simID].BookTitle
    custID_books = Book[Book['UserID'] == custID].BookTitle
    return set(simID_books) - set(custID_books)
```
