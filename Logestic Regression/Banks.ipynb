```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
```

```python
# Reading csv file and storing it in Banks
Banks =pd.read_csv("bank-full.csv",sep = ';')       
Banks
```

```python
#Converting categorical columns to numeric
labelencoder = LabelEncoder()
Banks.isetitem(1,labelencoder.fit_transform(Banks['job']))
Banks.isetitem(2,labelencoder.fit_transform(Banks['marital']))
Banks.isetitem(3,labelencoder.fit_transform(Banks['education']))
Banks.isetitem(4,labelencoder.fit_transform(Banks['default']))
Banks.isetitem(6,labelencoder.fit_transform(Banks['housing']))
Banks.isetitem(7,labelencoder.fit_transform(Banks['loan']))
Banks.isetitem(8,labelencoder.fit_transform(Banks['contact']))
Banks.isetitem(10,labelencoder.fit_transform(Banks['month']))
Banks.isetitem(15,labelencoder.fit_transform(Banks['poutcome']))
Banks.isetitem(16,labelencoder.fit_transform(Banks['y']))
Banks

# old Method
# Banks.iloc[:,1] = labelencoder.fit_transform(Banks.iloc[:,1])
# Banks.iloc[:,2] = labelencoder.fit_transform(Banks.iloc[:,2])
# Banks.iloc[:,3] = labelencoder.fit_transform(Banks.iloc[:,3])
# Banks.iloc[:,4] = labelencoder.fit_transform(Banks.iloc[:,4])
# Banks.iloc[:,6] = labelencoder.fit_transform(Banks.iloc[:,6])
# Banks.iloc[:,7] = labelencoder.fit_transform(Banks.iloc[:,7])
# Banks.iloc[:,8] = labelencoder.fit_transform(Banks.iloc[:,8])
# Banks.iloc[:,10] = labelencoder.fit_transform(Banks.iloc[:,10])
# Banks.iloc[:,15] = labelencoder.fit_transform(Banks.iloc[:,15])
# Banks.iloc[:,16] = labelencoder.fit_transform(Banks.iloc[:,16])
```

```python
# Defining X & Y (y is to be predicted)
X = Banks.iloc[:,:15]       #selecting all rows & clmn from index 0 to 15
Y = Banks.iloc[:,16]        #selecting all rows of clmn index no 16
```

```python
#Logistic regression and fit the model
classifier = LogisticRegression(max_iter=5000)   
classifier.fit(X,Y)                              #inserting X,Y in classifier
```

```python
#predicting x & storing in y_pred
y_pred = classifier.predict(X)       
y_pred                              #printing predicted val  
```

```python
#converting predicted arrays to df and adding actual clmn
y_pred_df= pd.DataFrame({'actual': Y,'predicted': classifier.predict(X)})    
y_pred_df.head()
```

```python
#Printing only different predictions
y_pred_df[y_pred_df["actual"] != y_pred_df["predicted"]]     
```

```python
# Gives count of client subscribed to term deposit or not    (0 = No,  1 = Yes)
y_pred_df["predicted"].value_counts()
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr, tpr, thresholds = roc_curve(Y, classifier.predict_proba (X)[:,1])
#drawing with Y & prob x clmn 1  1=true positive rate
auc = roc_auc_score(Y, y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logit model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

# Confusion Matrix

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y,y_pred)
print (confusion_matrix)
```

# Classification Report

```python
#Classification report                                                0=positive 1=negative
## Precision – What percent of your predictions were correct?       true0/true0+false0
## Recall – What percent of the positive cases did you catch?       Recall=seansitivity/TPR formula=true 0/true 0+false 1 
## F1 score – What percent of positive predictions were correct?    2*recall*precision/recall+precision
from sklearn.metrics import classification_report
print(classification_report(Y,y_pred))
```
